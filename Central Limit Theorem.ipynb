# Central Limit Theorem

The Central Limit Theorem (CLT) is a fundamental concept in statistics that describes the behavior of the distribution of sample means as the sample size increases, regardless of the shape of the population distribution. It states that the sampling distribution of the sample mean will be approximately normally distributed, regardless of the shape of the original population distribution, under certain conditions

We can test this by getting a poulation  of uniformly distributed data..

```python
import random
import seaborn as sns
import pandas as pd
from matplotlib import pyplot as plt
```

## a) Generating a Uniformly Distributed Population

```python
#We can start by generating uniform data
uniform_data = random.uniform(0, 1)
uniform_data = [round(random.uniform(0, 100)) for i in range (100000)]

data = pd.DataFrame(uniform_data, columns = ["Value"])
frequency_table = data["Value"].value_counts().reset_index()
frequency_table
```

```python
sns.kdeplot(uniform_data, fill=True)
plt.title("1. Plotting the PDF in our Uniform Distribution")
plt.xlabel("Outcomes")
```

## b) Getting the Means of Samples Drawn from the Uniform Population

```python
#Drawing random samples and finding their mean

sample_set = []
for i in range(1, 100):
    sample = []
    for i in range(1, 50):
        sample.append(random.choice(uniform_data))
    sample_mean = sum(sample)/len(sample)
    sample_set.append(sample_mean)
```

```python
sns.kdeplot(sample_set, fill=True)
plt.title("2. PDF of the Sample-Means Distribution")
```

## C) Results

On the second plot we get the distribution of the the sample means to approach a normal distribution even when the population the samples are drawn from are not normally distributed and are rather uniformly distributed.
